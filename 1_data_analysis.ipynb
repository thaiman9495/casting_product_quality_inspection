# Dataset distribution

## Train data

```python
path_train_defect_images = path_parent.joinpath("data/casting_data_300x300/train/def_front")
path_train_ok_images = path_parent.joinpath("data/casting_data_300x300/train/ok_front")
```

```python
n_train_defect_images = len([1 for x in path_train_defect_images.glob("*.jpeg")])
n_train_ok_images = len([1 for x in path_train_ok_images.glob("*.jpeg")])
```

```python
print(f"There are {n_train_ok_images} 'ok' images and {n_train_defect_images} 'defect' images in the train dataset")
```
